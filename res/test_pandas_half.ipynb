```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

im = np.array([[[x*3/4,127,y*3/4] for x in range(256)] for y in range(256)]).astype(np.uint8)
df = pd.DataFrame([[1,2,3],[4,5,6]], index=['AA','BB'], columns=['A','B','C'])
print(df)
fig = plt.figure(figsize=(8, 5), dpi=96)
ax = fig.subplots(1, 3)
ax[0].plot([1,2,3],[4,5,6])
ax[1].plot(df)
ax[2].imshow(im)
plt.show()
```
